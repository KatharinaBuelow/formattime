```python
import pandas as pd
import numpy as np
import xarray as xr

# Open the ERA5-Land dataset
era5land = xr.open_dataset(era5land)

# Extract the time values
time_values = era5land['time'].values

# Separate the integer date and fractional part
integer_part = np.floor(time_values).astype(int)  # Extract the date part (e.g., 19600921)
fractional_part = time_values - integer_part      # Extract the fractional part (e.g., 0.479167)

# Convert the integer part to datetime
dates = pd.to_datetime(integer_part.astype(str), format='%Y%m%d')

# Add the fractional part as timedelta
times = dates + pd.to_timedelta(fractional_part * 24, unit='h')  # Convert fractional day to hours

# Assign the converted times back to the dataset
era5land['time'] = times

# Verify the conversion
print(era5land['time'])
```
